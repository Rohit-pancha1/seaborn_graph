```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df= sns.load_dataset('tips')
```

```python
df.head()
```

```python
x = df['total_bill']
y = df['tip']
```

```python
sns.scatterplot(x='total_bill', y='tip',data=df, hue='time')
```

```python
sns.relplot(
    data=df,
    x = 'total_bill', y = 'tip',
    hue= 'time',
    size= 'size',
    sizes= (10,100),palette= 'Accent_r'
)
```

```python
import matplotlib.pyplot as plt
sns.scatterplot(
    data=df,
    x = 'total_bill', y = 'tip',
    hue= 'time',
    size= 'size',
    sizes= (10,100),palette= 'Accent_r'
)
plt.xlabel('Total Bill Generated')

plt.ylabel('Tips to the service in usd $')
plt.title('NEW York Reastourant')
```

```python
sns.lineplot(data=df,x='total_bill', y='size',hue='sex',style='smoker', markers= True, palette= 'Blues_d', dashes=True)
```
